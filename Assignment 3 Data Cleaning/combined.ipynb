```python
import pandas as pd
import numpy as np
# Pandas will automatically convert a column of integers to floats 
# if there are any missing values (NaN) in the column because NaN is considered a float value. This is because the numpy.nan value, which is used to represent missing data, is a special floating-point value.
#  remember we need to consider for future purpose: make the size bigger using common sense
path = prefix + "inventory.csv"
df = pd.read_csv(path)
# loop over each column
for col in df.columns:
    if df[col].dtype == 'object':  # if the column is of object (string) type
        max_length = df[col].str.len().max()
        print(f"Maximum string length in {col}: {max_length}")
    else:
        if np.issubdtype(df[col].dtype, np.number):  # if the column is of numeric type
            max_value = df[col].max()
            min_value = df[col].min()
            float_values = df[col][df[col].apply(lambda x: isinstance(x, float))]
            print(f"Maximum value in {col}: {max_value}")
            print(f"Minimum value in {col}: {min_value}")
            if not float_values.empty:
                print(f"Float values in {col}: {float_values.tolist()}")
```
