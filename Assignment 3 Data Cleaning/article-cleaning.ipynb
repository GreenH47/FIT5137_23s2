# data pre cleaning
we can not directly open the csv. In that case we need seperate line by line to locate and isolate the issue.

```python
article_file_path = 'rawdata/article.csv'
with open(article_file_path, 'r') as f:
    lines = f.readlines()

# Get the number of columns in the header
header_columns = len(lines[0].split(','))

# Separate the lines into correct and incorrect lines
correct_lines = []
incorrect_lines = []
incorrect_lines.append(lines[0])


# Check each line
for i, line in enumerate(lines):
    num_columns = len(line.split(','))
    if num_columns != header_columns:
        print(f'Row {i} has  {num_columns} of columns instead of {header_columns}')
        incorrect_lines.append(line)
    else:
        correct_lines.append(line)

# Write the correct lines to a new file
correct_file_path = 'format_data/article-correct-lines.csv'
with open(correct_file_path, 'w') as f:
    f.writelines(correct_lines)

# Write the incorrect lines to a new file
incorrect_file_path = 'format_data/article-incorrect-lines.csv'
with open(incorrect_file_path, 'w') as f:
    f.writelines(incorrect_lines)


```

After the data cleaning we can sperate the correct and incorrect lines in different files.

```python
# Calculate the count for each sex and categoryName
counts = df.groupby(['sex', 'categoryName']).size().reset_index(name='count')

# Calculate the total count for each sex
totals = df.groupby('sex').size()

# Calculate the percentage of each count relative to the total for that sex
counts['percentage'] = counts.apply(lambda row: round(row['count'] / totals[row['sex']] * 100, 4), axis=1)
# Sort the DataFrame by 'sex' and 'percentage'
counts = counts.sort_values(by=['sex', 'percentage'], ascending=[True, False])

print(counts)
# Save the counts DataFrame to a CSV file
counts.to_csv('format_data/articles_counts.csv', index=False)
```

## cleaning for the VendorKey and VendorName columns

```python
import pandas as pd
df = pd.read_csv('format_data/articles_final.csv')

distinct_combinations = df[['VendorKey', 'VendorName']].drop_duplicates()
distinct_combinations
print(distinct_combinations.groupby("VendorKey").count())

count = distinct_combinations.groupby("VendorKey").size()
print(count[count>1])

```

so we can find that there are duplicate "ANT" value inside VendorKey column. in that case we need find those rows with "ANT" value

```python
selected_row = distinct_combinations[distinct_combinations['VendorKey'] == 'ANT']

print(selected_row)
```

### VendorKey with ANT value cleaning

```python
selected_row = df[(df['VendorKey'] == 'ANT') & (df['VendorName'] == 'VIRIDIMIE')]
print(selected_row)
num_rows, num_columns = selected_row.shape
print(f'There are {num_rows} rows duplicated for ANT and VIRIDIMIE')
```

```python
selected_row = distinct_combinations[distinct_combinations['VendorName'] == 'VIRIDIMIE']

print(selected_row) 
```

we can find that the VendorName with VIRIDIMIE only have ANT as it VendorKey.   
So we need check another VendorName COOLROAM which also shared ANT as its VendorKey

```python
selected_row = df[(df['VendorKey'] == 'ANT') & (df['VendorName'] == 'COOLROAM')]
print(selected_row)
num_rows, num_columns = selected_row.shape
print(f'There are {num_rows} rows duplicated for ANT and COOLROAM')
```

so we need find out the COOLROAM most used VendorKey besides ANT

```python
distinct_combinations = df[['VendorKey', 'VendorName']].drop_duplicates()
selected_row = distinct_combinations[distinct_combinations['VendorName'] == 'COOLROAM']

count = df[df['VendorName'] == 'COOLROAM'].groupby('VendorKey').size().reset_index(name='count')
selected_row = selected_row.merge(count, on='VendorKey')

print(selected_row)
```

in here we can find that COOLROAM used most of CLR as its VendorKey. So we need change all COOLROAM with CLR VendorKey.

```python
# Select rows where VendorName is 'COOLROAM' and change VendorKey to 'CLR'
df.loc[df['VendorName'] == 'COOLROAM', 'VendorKey'] = 'CLR'

distinct_combinations = df[['VendorKey', 'VendorName']].drop_duplicates()
selected_row = distinct_combinations[distinct_combinations['VendorName'] == 'COOLROAM']

count = df[df['VendorName'] == 'COOLROAM'].groupby('VendorKey').size().reset_index(name='count')
selected_row = selected_row.merge(count, on='VendorKey')

print(selected_row)

# Save the updated DataFrame to format_data/articles_final.csv
df.to_csv('format_data/articles_final.csv', index=False)
```
